```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori

#Data 불러오기
roughResult = pd.read_csv('BI_data_new.csv')
tempFashionResult = roughResult.iloc[:,[1,2]]
```

```python
#MusicData 정제
result3 = []
for i in range(len(tempFashionResult)):
    result1 = tempFashionResult.iloc[i, 0].split(',')
    for j in range(len(result1)):
        result2 = result1[j].split('(')
        for k in result2:
            if k != " " and k[-1] != ")" and k !=' Instrumental':
                result3.append(k.replace(" ", ""))

                
resultMusic = [["" for col in range(3)] for row in range(172)]
for i in range(len(result3)):
    row = int(i/3)
    col = i%3
    resultMusic[row][col] = result3[i]

resultMusic
```

```python
tempFashionResult
```

```python
#FashionData 정제
resultF = []
resultF1 = []
resultF4 = []
for i in range(len(tempFashionResult)):
    resultF1.append(tempFashionResult.iloc[i, 1])

for i in range(len(resultF1)):
    resultF1[i] = resultF1[i].replace("Workwear (워크웨어륵) (constructed of sturdy fabrics like denim and corduroy, and it has an oversized fit to allow for movement)", "Workwear")

for i in range(len(resultF1)):
    resultF2 = resultF1[i].split(',')
    for j in range(len(resultF2)):
        resultF3 = resultF2[j].split('(')
        for k in resultF3:
            if k != " " and k[-1] != ")" and k !=' Instrumental':
                resultF4.append(k.replace(" ", ""))
            
resultFashion = [["" for col in range(3)] for row in range(172)]
for i in range(len(resultF4)):
    row = int(i/3)
    col = i%3
    resultFashion[row][col] = resultF4[i]

resultFashion
```

```python
#popMusic table
popMusic = []
for i in range(len(resultMusic)):
    if 'Pop' in resultMusic[i]:
        popMusic.append(resultFashion[i])
popMusic
```

```python
#idolMusic table
idolMusic = []
for i in range(len(resultMusic)):
    if 'Idol' in resultMusic[i]:
        idolMusic.append(resultFashion[i])
idolMusic
```

```python
#BalladMusic table
balladMusic = []
for i in range(len(resultMusic)):
    if 'Ballad' in resultMusic[i]:
        balladMusic.append(resultFashion[i])
balladMusic
```

```python
#DanceMusic table
danceMusic = []
for i in range(len(resultMusic)):
    if 'Dance' in resultMusic[i]:
        danceMusic.append(resultFashion[i])
danceMusic
```

```python
#Rap/Hiphop Music table
rap_hiphopMusic = []
for i in range(len(resultMusic)):
    if 'Rap/HipHop' in resultMusic[i]:
        rap_hiphopMusic.append(resultFashion[i])
rap_hiphopMusic
```

```python
#R&B/Soul Music table
rb_soulMusic = []
for i in range(len(resultMusic)):
    if 'R&B/Soul' in resultMusic[i]:
        rb_soulMusic.append(resultFashion[i])
rb_soulMusic
```

```python
#Indie Music table
indieMusic = []
for i in range(len(resultMusic)):
    if 'Indie' in resultMusic[i]:
        indieMusic.append(resultFashion[i])
indieMusic
```

```python
#Rock/Metal Music table
rock_metalMusic = []
for i in range(len(resultMusic)):
    if 'Rock/Metal' in resultMusic[i]:
        rock_metalMusic.append(resultFashion[i])
rock_metalMusic

#트로트는 그냥 버렸어
```

```python
#popMusic Apriori
te = TransactionEncoder()
te_ary = te.fit(popMusic).transform(popMusic)
concluded = pd.DataFrame(te_ary, columns=te.columns_)
tempApriori = apriori(concluded, min_support=0.3, use_colnames=True)

popApriori = tempApriori.sort_values(["support"], ascending=[False])
popApriori
```
